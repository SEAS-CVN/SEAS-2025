```python
import scipy.io as sio
from scipy import fftpack as fft
import numpy as np 
import matplotlib.pyplot as plt
```

```python
H__sample_fu = np.load('./data1d/test/H__sample_fu.npy')
H_noisy__sample_fu = np.load('./data1d/test/H_noisy__sample_fu.npy')
print(H__sample_fu.shape, H_noisy__sample_fu.shape)
```

```python
sample_index = np.random.randint(H__sample_fu.shape[0])

channel__fu = H__sample_fu[sample_index]
channel__delay = fft.idct(channel__fu, axis=0)

noisy_channel__fu = H_noisy__sample_fu[sample_index]
noisy_channel__delay = fft.idct(noisy_channel__fu, axis=0)

delays = np.arange(0, channel__fu.shape[0], 1)

plt.plot(delays, 10*np.log10(np.abs(channel__delay)), label='True channel')
plt.plot(delays, 10*np.log10(np.abs(noisy_channel__delay)), label='Noisy channel')

plt.gca().set_aspect('equal')
plt.legend()
plt.show()
```
